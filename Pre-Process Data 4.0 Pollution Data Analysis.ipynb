```python
import pandas as pd 
import numpy as np
```

```python
# Read Pollution Data 
Pollution_pd = pd.read_csv('C:\\Users\\rarora17\\Desktop\\\Case_Studies\\Hackathon\\data\\AirPollutionData\\uspollution_pollution_us_2000_2016.csv')

Pollution_Data =  Pollution_pd.drop(['Unnamed: 0','City', 'State Code', 'County Code', 'Site Num', 'Address', 'County'],axis=1)
Pollution_Data.head()
```

```python
Pollution_Data =  Pollution_pd.drop(['Unnamed: 0','City', 'State Code', 'County Code', 'Site Num', 'Address', 'County'],axis=1)

Pollution_Data.columns

# Drop not to be used columns

Pollution_Data_Daily = Pollution_Data.drop(['NO2 Units', 'NO2 1st Max Value',
       'NO2 1st Max Hour', 'NO2 AQI', 'O3 Units', 'O3 AQI', 'SO2 Units',
       'SO2 1st Max Value', 'SO2 1st Max Hour', 'SO2 AQI',
       'O3 1st Max Value', 'O3 1st Max Hour',
       'CO Units', 'CO 1st Max Value', 'CO 1st Max Hour', 'CO AQI'],axis=1)
Pollution_Data_Daily.head()


# Add 'Year' to the DateFrame 
Pollution_Data_Daily['Year'] = Pollution_Data['Date Local'].str.extract('(....)', expand=True)

Pollution_Data_Daily.head()

# Extract data for Years 2011 to 2016

#Convert 'Year' column type to int 
Pollution_Data_Daily['Year']= Pollution_Data_Daily.Year.apply(int)
Pollution_Data_Daily = Pollution_Data_Daily[Pollution_Data_Daily['Year'] > 2010]


Pollution_Data_Daily['Year'].value_counts()

#Pollution_Data_Daily = Pollution_Data_Daily[Pollution_Data_Daily['Year'] > 2010]
SO2Data = Pollution_Data_Daily.groupby(['Year','State'])['SO2 Mean'].mean()#.apply(lambda x: np.mean(x))
SO2Data.head()

SO2Data_df = pd.DataFrame(SO2Data)

SO2Data_df.sort_index(level=1).head()

```
